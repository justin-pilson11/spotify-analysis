```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
users = pd.read_csv('data/Users.csv')
churn = pd.read_csv('data/Churn.csv')
daily_listening_activity = pd.read_csv('data/DailyListeningActivity.csv')
feature_exposure = pd.read_csv('data/FeatureExposure.csv')

feature_exposure = feature_exposure[['user_id', 'exposed']]

subscriptions = pd.read_csv('data/Subscriptions.csv')
```

```python
daily_listening_activity.count()
```

```python
import sqlite3

conn = sqlite3.connect("analysis.db")

users.to_sql("users", conn, if_exists="replace", index=False)
daily_listening_activity.to_sql("daily_listening_activity", conn, if_exists="replace", index=False)
feature_exposure.to_sql("feature_exposure", conn, if_exists="replace", index=False)
subscriptions.to_sql("subscriptions", conn, if_exists="replace", index=False)
churn.to_sql("churn_labels", conn, if_exists="replace", index=False)
```
