```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
import sqlite3
```

```python
users = pd.read_csv('data/Users.csv')
daily_listening_activity = pd.read_csv('data/DailyListeningActivity.csv')

feature_exposure = pd.read_csv('data/FeatureExposure.csv')

subscriptions = pd.read_csv('data/Subscriptions.csv')
```

```python
import sqlite3

conn = sqlite3.connect("analysis.db")

users.to_sql("users", conn, if_exists="replace", index=False)
daily_listening_activity.to_sql("daily_listening_activity", conn, if_exists="replace", index=False)
feature_exposure.to_sql("feature_exposure", conn, if_exists="replace", index=False)
subscriptions.to_sql("subscriptions", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT COUNT(*) FROM users;", conn)
pd.read_sql("SELECT COUNT(*) FROM feature_exposure;", conn)
pd.read_sql("SELECT COUNT(*) FROM daily_listening_activity;", conn)
pd.read_sql("SELECT COUNT(*) FROM subscriptions;", conn)
```

```python
pd.read_sql("PRAGMA table_info(users);", conn)
pd.read_sql("PRAGMA table_info(feature_exposure);", conn)
pd.read_sql("PRAGMA table_info(subscriptions);", conn)
pd.read_sql("PRAGMA table_info(daily_listening_activity);", conn)
```

```python
pd.read_sql('SELECT * FROM users LIMIT 5;', conn)
pd.read_sql('SELECT * FROM feature_exposure LIMIT 5;', conn)
pd.read_sql('SELECT * FROM subscriptions LIMIT 5;', conn)
pd.read_sql('SELECT * FROM daily_listening_activity LIMIT 5;', conn)
```

```python
pd.read_sql("""
SELECT
    exposed,
    COUNT(DISTINCT user_id) AS users
FROM feature_exposure
GROUP BY exposed;
""", conn)
```
